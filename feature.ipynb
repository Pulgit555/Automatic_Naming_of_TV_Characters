```python
import cv2
import dlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("./frames/frame1745.jpg")
detector = dlib.get_frontal_face_detector()
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = detector(gray)
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")
points = [17, 21, 22, 26, 30, 31, 35, 48, 54]
for face in faces:
    x1 = face.left() # left point
    y1 = face.top() # top point
    x2 = face.right() # right point
    y2 = face.bottom() # bottom point
    img = cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 1)

    landmarks = predictor(gray, face)
    for point in points:
        x = landmarks.part(point).x
        y = landmarks.part(point).y
        img = cv2.circle(img, (x, y), 2, (0, 255, 0), -1)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# read images
img1 = cv2.imread('joey1.jpg')  
img2 = cv2.imread('joey2.jpg') 

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

#sift
sift = cv2.xfeatures2d.SIFT_create()

keypoints_1, descriptors_1 = sift.detectAndCompute(img1,None)
keypoints_2, descriptors_2 = sift.detectAndCompute(img2,None)

#feature matching
bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)

matches = bf.match(descriptors_1,descriptors_2)
matches = sorted(matches, key = lambda x:x.distance)
print(len(matches))
img3 = cv2.drawMatches(img1, keypoints_1, img2, keypoints_2, matches[:50], img2, flags=2)
plt.imshow(img3),plt.show()
```

```python
img = cv2.imread("./frames/frame2558.jpg")
img1 = cv2.imread('./images/monica.jpg')
img1 = cv2.resize(img1, (img.shape[1], img.shape[0]), interpolation=cv2.INTER_CUBIC)
detector = dlib.get_frontal_face_detector()
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
faces = detector(gray)
faces1 = detector(gray1)
mask = np.zeros(gray.shape, dtype="uint8")
mask1 = np.zeros(gray1.shape, dtype="uint8")
for face in faces:
    x1 = face.left() # left point
    y1 = face.top() # top point
    x2 = face.right() # right point
    y2 = face.bottom() # bottom point
    img = cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 1)
    for i in range(mask.shape[0]):
        for j in range(mask.shape[1]):
            if i > y1 and i < y2 and j > x1 and j < x2:
                mask[i][j] = 1
    keypoints_1, descriptors_1 = sift.detectAndCompute(gray,mask=mask)
for face in faces1:
    x1 = face.left() # left point
    y1 = face.top() # top point
    x2 = face.right() # right point
    y2 = face.bottom() # bottom point
    img1 = cv2.rectangle(img1, (x1, y1), (x2, y2), (0, 255, 0), 1)
    for i in range(mask1.shape[0]):
        for j in range(mask1.shape[1]):
            if i > y1 and i < y2 and j > x1 and j < x2:
                mask1[i][j] = 1
    keypoints_2, descriptors_2 = sift.detectAndCompute(gray1,mask=mask1)
print(descriptors_1.shape)
print(descriptors_2.shape)
# img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)
matches = bf.match(descriptors_1,descriptors_2)
matches = sorted(matches, key = lambda x:x.distance)
print(len(matches))
img3 = cv2.drawMatches(img, keypoints_1, img1, keypoints_2, matches[:50], img1, flags=2)
plt.imshow(img3),plt.show()
```
